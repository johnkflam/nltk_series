```python
def tag_all_notes(df):
    COLUMNS = ['mon_note','tue_note','wed_note','thu_note','fri_note']
    tagger = nu.get_regexp_tagger()
    tagged_list = []
    regex_tokenizer = nu.get_regex_tokenizer()
    #sw = nu.get_stopwords(['none','concur',',','.','&','-','(',')','/',';',':','#','+'])
    for idx, row in df.iterrows():
        for col in COLUMNS:
            hours = col[:4]+'hrs'
            #words = nltk.word_tokenize(row[col])
            words = regex_tokenizer.tokenize(row[col])
            tagged_note = tagger.tag(words)
            rec = (row['pj_id'],col,tagged_note,row[hours])
            tagged_list.append(rec)
    #return nltk.pos_tag(all_words)
    
    return tagged_list
```
